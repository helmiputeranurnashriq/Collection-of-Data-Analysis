```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

Hi, here is my first try websrapping python. 

```python
# request engine
page = requests.get('https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2380057.m570.l1311&_nkw=graphics+card&_sacat=0')
soup = BeautifulSoup(page.content, 'html.parser')
data_list = soup.find(class_ = 'srp-river srp-layout-inner')
data_info = data_list.find_all(class_ = 's-item__wrapper clearfix')

df =pd.DataFrame
lists = [] # create a list
listsPrice = [] # create a list

# webscrapping engine
for item in data_info:
    item_name = item.find(class_ = 's-item__title').get_text()
    lists.append(item_name)

    item_price = item.find(class_ = 's-item__detail s-item__detail--primary').get_text()

    listsPrice.append(item_price)
   
   
# Creating pandas dataframe
df = pd.DataFrame({'Item Name':lists, 'Price (in Dolar,$)': listsPrice})
df = df.drop([0])
df_copy = df.copy()
df_copy
```

I wanted to create a new column for item model either AMD or NVIDIA. So I had used lambda to identify those words in item name. If true then It will return true value.

```python
#Lambda Function. 
#df_copy[df_copy['Item Name'].str.contains('AMD')]
df_copy["Item Model"] = df_copy['Item Name'].apply(lambda x: "AMD" if "AMD" in x else ("NVIDIA" if "NVIDIA" in x else ("AMD" if "Radeon" in x else "NVIDIA")))
df_copy
```

Here is data cleaning process. I removed any unwanted strings.

```python
#Data Cleaning
df_copy["Item Name"] = df_copy["Item Name"].str.replace('New Listing', '')
df_copy["Item Name"] = df_copy["Item Name"].str.replace('ðŸ”¥', '')
df_copy["Item Name"] = df_copy["Item Name"].str.replace('âœ… New âœ…', '')
df_copy["Item Name"] = df_copy["Item Name"].str.replace('"Benchmark Tested"', '')
df_copy
```

Again, I wanted to create another a new column for GPU engine. So I used lambda function.

```python
#Lambda Function
df_copy["GPU Engine"] = df_copy['Item Name'].apply(lambda x: "GAMING" if "RX" in x else ("GAMING" if "RTX" in x else( "GAMING" if "GTX" in x else"WORK STATION/LOW PERFORMANCE CARD/OLD GEN")))
df_copy
```

Below code was used by me to identify brand name. After I used split method with expand = true, I found brand name is in the 1st column. Then I used this opportunity to create a new column brand name.

```python
# Data manipulation
df_copy["Item Name"] = df_copy["Item Name"].str.lstrip()
df_copy["Brand"] = df_copy["Item Name"].str.rsplit(' ',expand = True)[0]
df_modified = df_copy.copy()
```

```python
#Data Cleaning
df_modified["Price (in Dolar,$)"] = df_modified["Price (in Dolar,$)"].str.replace('$','').str.strip()
df_modified["Price (in Dolar,$)"] = df_modified["Price (in Dolar,$)"].str.replace(',', '')
df_modified
```

```python
# Data Cleaning process
df_modified["Price (in Dolar,$)"] = df_modified["Price (in Dolar,$)"].astype(float)
df_modified["Price (RM)"] = (df_modified["Price (in Dolar,$)"]*4.19).round(2)
df_modified["Price (RM)"] = "RM "+ df_modified["Price (RM)"].astype(str)
df_modified["Price (in Dolar,$)"] = "$" + df_modified["Price (in Dolar,$)"].astype(str)
df_modified
```
